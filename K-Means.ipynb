## KMeans with 5 clusters 

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=5, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(df[['BALANCE','PURCHASES']]) #we choose the x and y we are going to work on and we fit them to the model 
```

```python
labels=pd.DataFrame(kmeans.labels_)
df["label"]=labels    #check our clusters labels 
```

```python
# df["label"].groupby('label').count
df.groupby(['label']).count()
```

```python
plt.scatter(df["BALANCE"][df["label"]==0], #'[df["label"]==0]' ,with 0 we mean the index of the cluster         
            df["PURCHASES"][df["label"]==0],s=80,c='magenta',label='cluster1')
plt.scatter(df["BALANCE"][df["label"]==1],
           df["PURCHASES"][df["label"]==1],s=80,c='yellow',label='cluster2')
plt.scatter(df["BALANCE"][df["label"]==2],
           df["PURCHASES"][df["label"]==2],s=80,c='green',label='cluster3')
plt.scatter(df["BALANCE"][df["label"]==3], 
           df["PURCHASES"][df["label"]==3],s=80,c='cyan',label='cluster4')
plt.scatter(df["BALANCE"][df["label"]==4], 
           df["PURCHASES"][df["label"]==4],s=80,c='burlywood',label='cluster5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
#since we only selected two features to work on as x and y (we could have used PCA to gather many features to use as x and y)
#so 'kmeans.cluster_centers_[:, 0]' the 0 is the index of the BALANCE feature in out new dataframe 
#so 'kmeans.cluster_centers_[:, 1]' the 1 is the index of the PURCHASES feature in out new dataframe 

plt.title('Clusters of Customers')
plt.xlabel('BALANCE')
plt.ylabel('PURCHASES')
plt.legend()
plt.show()
```

## Elbow Method

```python
Sum_of_squared_distances = []
K = range(1,15)
for k in K:
    km=KMeans(n_clusters=k)
    km=km.fit(df)
    Sum_of_squared_distances.append(km.inertia_)
    
```

```python
plt.plot(K, Sum_of_squared_distances,'bx-')
plt.xlabel('k')
plt.ylabel('Sum_of_squared_distances')
plt.title('Elbow Method')
plt.show()
```

## KMeans with 4 clusters

```python
from sklearn.cluster import KMeans  #Importing our clustering algorithm: KMeans
kmeans=KMeans(n_clusters=4, random_state=0)  #Cluster our data by choosing 5 as number of clusters
kmeans.fit(df[['BALANCE','PURCHASES']])
```
