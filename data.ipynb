```python
def plot_correlation_map( df ):

    corr = df.corr()

    s , ax = plt.subplots( figsize =( 12 , 10 ) )

    cmap = sns.diverging_palette( 220 , 10 , as_cmap = True )

    s = sns.heatmap(

        corr, 

        cmap = cmap,

        square=True, 

        cbar_kws={ 'shrink' : .9 }, 

        ax=ax, 

        annot = True, 

        annot_kws = { 'fontsize' : 12 }

        )
```

```python
plot_correlation_map(df)
```

The plot_correlation_map function creates a correlation matrix that measures the correlation between each pair of features on a scale from -0.4 to 1.0 (1.0 for a perfect correlation),that's why the diagonal values are all 1.0 because it's a correlation between the same feature,and the matrix is diagonal
